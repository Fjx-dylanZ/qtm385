```python
import random as rd

KEY = 100
ABC = 'abcdefghijklmnopqrstuvwxyz'
message = ''
usr_command = ''

def fileCheck():
  global message
  try:
    message = open('message.txt', 'r')
  except:
    print('Error. \'message.txt\' is not availabe')
    return False
  else:
    return True

def cypher(mode, mapOnly, echo):
  global message
  fileByChar = []

  if fileCheck():
    rd.seed(KEY)
    encrypted_ABC = rd.sample(ABC, 26)
    if mode == 'encrypt':
      cryptic_map = dict(zip(ABC, encrypted_ABC))
    else:
      cryptic_map = dict(zip(encrypted_ABC, ABC))

    if mapOnly:
      for char in cryptic_map:
        print("{key} => {value}".format(key = char, value = cryptic_map.get(char)),
              end = "\t"
              )
      print("\n")
      return 0
    else:
      while True:
        char = message.read(1)          
        if not char:
          break
        fileByChar.append(char)

      for count, char in enumerate(fileByChar):
        if (char in ABC) or (char in ABC.upper()):
          if char.islower():
            fileByChar[count] = cryptic_map.get(char)
          if char.isupper():
            fileByChar[count] = cryptic_map.get(char.lower()).upper()
      if not echo:

        if mode == 'encrypt':
          print("Successful encryption.")
        else:
          print("Successfully decrypted.")
          print("Your message is: ")
          cypher('decrypt', False, True)
            
        #Delete Original Texts
        message = open('message.txt', 'w')
        message.write('')
        message.close()
        
        #Write
        message = open('message.txt', 'a')
        for char in fileByChar:
          message.write(char)
        message.close()


      else:
        lineList = []
        temp = ''
        for char in fileByChar:
          if char != "\n":
            temp += char
          else:
            temp += "\n"
            lineList.append(temp)
            temp = ''
        if temp != '':
          lineList.append(temp)
        show_message(lineList)
  else:
    if mode == 'encrypt':
      print("Error. The encryption failed.")
    else:
      print("Error. The decryption failed.")


def set_key():
  global KEY
  try:
    temp = int(input("Type the encryption key: "))
  except: 
    print("Error: it should be a number between 100 and 999")
  else:
    if temp in range(100,1000):
      KEY = temp
    else:
      print("Error: it should be a number between 100 and 999")


def show_message(lines):
  if type(lines) is not list:
    if fileCheck():
      lines = message.readlines()
      message.close()
  if fileCheck():
    if len(lines) < 11:
      for line in lines:
        print(line, end = '')
    else:
      for i in range(0,5):
        print(lines[i], end = '')
      print('...')
      for i in range(len(lines)-5, len(lines)):
        print(lines[i], end = '')



while(usr_command != 'bye'):
  usr_command = input(">> ")
  if usr_command == 'bye':
    print('Bye! Have a nice day.')

  elif usr_command == 'encrypt':
    cypher('encrypt', False, False)

  elif usr_command == 'decrypt':
    cypher('decrypt', False, False)
  
  elif usr_command == 'show-message':
    show_message(2)
  
  elif usr_command == 'show-encrypted':
    cypher('encrypt', False, True)

  elif usr_command == 'show-decrypted':
    cypher('decrypt', False, True)
  
  elif usr_command == 'show-map':
    print("The encryption map is: ")
    cypher('encrypt', True, False)
    print("And the decruption map is: ")
    cypher('decrypt', True, False)
  
  elif usr_command == 'set-key':
    set_key()

  elif usr_command == 'show-key':
    print('The encryption key is: {}'.format(KEY))

  else:
    print("Eorror. Invalid command.")
```
